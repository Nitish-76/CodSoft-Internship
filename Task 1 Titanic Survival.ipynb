## 3.2 Coding categorical values.

```python
# Coding the embarked places as 0=Q, 1=S and 2=C.
df1['Embarked'].replace({'Q':'0','S':'1','C':'2'}, inplace= True)
```

## 3.3 Seperating the target and feature columns

```python
# Seperating the target and feature columns
X = df1.drop(columns = ['PassengerId', 'Name', 'Ticket', 'Survived', 'Fare','Cabin'], axis=1)
y = df1['Survived']
```

## 3.4 Splitting the data into testing and training.

```python
!pip install -U scikit-learn
!pip install scipy
!pip install joblib
!pip install threadpoolctl
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
```

```python
# 20% is test data and 80% is training data.
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=2)
print(X.shape, X_train.shape,X_test.shape)
```

```python
X_train.dtypes
```

# 4. Training and testing

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear', random_state = 0)
```

```python
# Loading the model
model.fit(X_train,y_train)
```

```python
#ON train data
Give = model.predict(X_train)
```

```python
print(Give)
```

```python
# Accuracy score on train data.
accur_score = accuracy_score(y_train,Give)
print("Accuracy score of the model is:",accur_score)
```

```python
#On test data
Test = model.predict(X_test)
print(Test)
```

```python
# Accuracy score on test data
acc_score = accuracy_score(y_test,Test)
print("Accuracy score of the model:",acc_score)
```
